```python
import os, glob

BASE = "/content/drive/MyDrive/Base de Dados"

print("Existe BASE?", os.path.exists(BASE))
print("Conteúdo BASE:", os.listdir(BASE))

# Garantir que as 3 pastas existam
for sub in ["Treino", "Teste", "Oficial"]:
    p = os.path.join(BASE, sub)
    os.makedirs(p, exist_ok=True)

print("\nDepois de garantir as pastas:")
print(sorted(os.listdir(BASE)))

# Listar conteúdo de cada pasta
for sub in ["Treino", "Teste", "Oficial"]:
    p = os.path.join(BASE, sub)
    print("\n=== Pasta:", sub, "===")
    print("Caminho:", p)
    print("Arquivos:", os.listdir(p))
    print("ZIPs encontrados:", glob.glob(os.path.join(p, "*.zip")))
```
